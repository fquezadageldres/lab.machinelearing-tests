```python
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

model = load_model('model_p003_3classes.h5')


image_path = "test/999.jpg"
img_ = image.load_img(image_path, target_size=(128, 128))
img = np.expand_dims(img_, axis=0)
img = img/255

result = model.predict(img)

print('Cat: '+ str("%.2f" % (result[0][0]*100)) + '%\nDog: '+ str("%.2f" % (result[0][1]*100)) + '%\nOther: '+ str("%.2f" % (result[0][2]*100)) + '%')
plt.imshow(img_)
plt.show()
```
